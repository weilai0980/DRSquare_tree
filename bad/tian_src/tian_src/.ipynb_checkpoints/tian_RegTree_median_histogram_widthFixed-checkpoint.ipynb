```python
# basic functions in one-pass method [on the cluster side]

# assign each data instance to the node
def search_nodeToData(features, tree):
    nodeNum=len(tree)   
    if nodeNum == 0:
        return 0;
    
    current_nodeIdx=0
    while current_nodeIdx< nodeNum:
        split_feature= tree[current_nodeIdx][0]
        split_valueSet= tree[current_nodeIdx][1]
        
        if split_feature==-1:
            return -1
    
        if (( 1<<features[split_feature]) &  split_valueSet) != 0 :
            current_nodeIdx=current_nodeIdx*2+1
        else:
            current_nodeIdx=current_nodeIdx*2+2        
    return current_nodeIdx - nodeNum


def partition_combiner_hist(list_dvAndfeatures):
    nodes_dict={}
    for dvAndfeatures in list_dvAndfeatures:
        
        Y= dvAndfeatures[0]
        node= search_nodeToData(dvAndfeatures[1:numFeatures+1], \
                                currentNode_split_fromMaster.value)
        # equi-width
        binId=math.floor(1.0* Y/bin_width )
        
        if node in nodes_dict.keys():
            
            # new added: sum of y in a node
            nodes_dict[node]['sumY'] = nodes_dict[node]['sumY']+Y 
            nodes_dict[node]['count'] = nodes_dict[node]['count']+1 
            
            if binId in nodes_dict[node]['hist'].keys():
                nodes_dict[node]['hist'][binId]= nodes_dict[node]['hist'][binId]+1
            else:
                nodes_dict[node]['hist'].update( {binId:1}  )
          
        
            for i in range(0,numFeatures):        
                feature_val= dvAndfeatures[i+1]
                
                if feature_val not in nodes_dict[node][i]:
                    nodes_dict[node][i].update( { feature_val:{} }  )
                    nodes_dict[node][i][feature_val].update( {'count_inFeatureValue':1} )
                    nodes_dict[node][i][feature_val].update( {'hist_inFeatureValue':{}} )
                    nodes_dict[node][i][feature_val]['hist_inFeatureValue'].update({binId:1} )
                    
                else:
                    nodes_dict[node][i][feature_val]['count_inFeatureValue']= \
                    nodes_dict[node][i][feature_val]['count_inFeatureValue']+1
                    
                    if binId not in nodes_dict[node][i][feature_val]['hist_inFeatureValue'].keys():
                        nodes_dict[node][i][feature_val]['hist_inFeatureValue'].update( {binId:1} )
                    else:
                        nodes_dict[node][i][feature_val]['hist_inFeatureValue'][binId]= \
                        nodes_dict[node][i][feature_val]['hist_inFeatureValue'][binId]+1
                        
      
        else:
            nodes_dict.update( {node: {}})  
            # new added: sum of y in a node
            nodes_dict[node].update( { 'sumY': Y} )
            nodes_dict[node].update( { 'count': 1} )
            nodes_dict[node].update( { 'hist': {}} )
            nodes_dict[node]['hist'].update( {binId:1} )
                
            for i in range(0,numFeatures):
                feature_val= dvAndfeatures[i+1]
                
                nodes_dict[node].update( { i: {}} )    
                nodes_dict[node][i].update( { feature_val: {}   }  )
                
                nodes_dict[node][i][feature_val].update( {'count_inFeatureValue':1} )
                nodes_dict[node][i][feature_val].update( {'hist_inFeatureValue':{}} )
                nodes_dict[node][i][feature_val]['hist_inFeatureValue'].update( {binId:1} )
                
    return zip(nodes_dict.keys(), nodes_dict.values())

def merge_parttion_combiner_hist(nodeToFeatureToValue_1,nodeToFeatureToValue_2 ):
# optimization: calculate median and amd for feature-value    
    
    # new added: sum of Y in a node
    nodeToFeatureToValue_1['sumY']= nodeToFeatureToValue_1['sumY']+\
    nodeToFeatureToValue_2['sumY']
    nodeToFeatureToValue_1['count']= nodeToFeatureToValue_1['count']+\
    nodeToFeatureToValue_2['count'] 
    
    for i in nodeToFeatureToValue_2['hist'].keys():
        if i in nodeToFeatureToValue_1['hist'].keys():
            nodeToFeatureToValue_1['hist'][i]= nodeToFeatureToValue_2['hist'][i]+\
            nodeToFeatureToValue_1['hist'][i]
        else:
            nodeToFeatureToValue_1['hist'].update( {i: nodeToFeatureToValue_2['hist'][i]}       )
            
    for i in range(0, numFeatures): #feature
        for j in nodeToFeatureToValue_1[i].keys(): #feature value
            
            feature_val=j
            
            if feature_val in nodeToFeatureToValue_2[i].keys():
                
                nodeToFeatureToValue_1[i][j]['count_inFeatureValue']=nodeToFeatureToValue_1[i][j]['count_inFeatureValue']+\
                nodeToFeatureToValue_2[i][j]['count_inFeatureValue']
                
                for k in nodeToFeatureToValue_2[i][j]['hist_inFeatureValue'].keys():
                    if k in nodeToFeatureToValue_1[i][j]['hist_inFeatureValue'].keys():
                        nodeToFeatureToValue_1[i][j]['hist_inFeatureValue'][k] = \
                        nodeToFeatureToValue_2[i][j]['hist_inFeatureValue'][k]+nodeToFeatureToValue_1[i][j]['hist_inFeatureValue'][k]
                    else:
                        nodeToFeatureToValue_1[i][j]['hist_inFeatureValue'].update({k:nodeToFeatureToValue_2[i][j]['hist_inFeatureValue'][k]})
                            
    for i in range(0, numFeatures):
        for j in nodeToFeatureToValue_2[i].keys():
            
            feature_val=j   
            if feature_val not in nodeToFeatureToValue_1[i].keys():
                nodeToFeatureToValue_1[i].update({feature_val: {} })
                nodeToFeatureToValue_1[i][feature_val]=nodeToFeatureToValue_2[i][feature_val].copy()
                
    return  nodeToFeatureToValue_1
```

```python
# split decision [on the local side]

#implementation notes:
# bit-operation to record the set of selected feature values in the left child

def hist_minus( hist1, hist2):
#hist1-hist2

    hist = copy.deepcopy(hist1)
    for i in hist2.keys():
        hist[i]= hist[i]-hist2[i]
    return hist


def hist_merge( hist1, hist2):
    
    hist= copy.deepcopy(hist1)  
    for i in hist.keys():
        if i in hist2.keys():
            hist[i]= hist[i]+hist2[i]
    for i in hist2.keys():
        if i not in hist.keys():
            hist.update( {i:hist2[i]} )
    return hist
 
def hist_meanAbsoluteDevi( hist, cnt ):
    
    med_cnt = cnt*1.0/2
    med_bin=-1
    med_bin_idx=-1
    
    sorted_hist = sorted(hist.items(), key= lambda val: val[0] ) #??? could be optimized
    numBins= len(sorted_hist)
    
    cntSum=0
    binByCnt=0
    tmp_cntsum=0
    
    cntSum_beforeMid=0
    
    #debug
#     beforeMed=[]
#     afterMed=[]
  
    for i in range(0,numBins):
        
        tmp_cntsum=cntSum + sorted_hist[i][1]
        
#         print tmp_cntsum,cntSum, med_cnt
        
        if med_bin_idx == -1:
            
            if tmp_cntsum*1.0 >= med_cnt:
                
                cntSum_beforeMid = cntSum
                
                med_bin= sorted_hist[i][0]
                med_bin_idx=i
                
                #debug
#                 beforeMed.append( cntSum)
#                 beforeMed.append(binByCnt)
                
            else:
                cntSum= tmp_cntsum 
#                 binByCnt= binByCnt - 1.0*sorted_hist[i][1]*(sorted_hist[i][0]+0.5)
                binByCnt= binByCnt - 1.0*sorted_hist[i][1]*(sorted_hist[i][0]+0.5*bin_width)

        
        else:
            
            cntSum= cntSum - sorted_hist[i][1]
#             binByCnt= binByCnt + 1.0*sorted_hist[i][1]*(sorted_hist[i][0]+0.5)
            binByCnt= binByCnt + 1.0*sorted_hist[i][1]*(sorted_hist[i][0]+0.5*bin_width)

    
    if cnt == 1:
        return ( (bin_width* (sorted_hist[i][0]*2+1)*0.5)  ,0)
    elif cnt ==0:
        return (0,0)
    else:        
        
        sample_inMedBin = med_cnt - cntSum_beforeMid
        
        med= bin_width*med_bin + bin_width*1.0/(sorted_hist[med_bin_idx][1]+1.0)*sample_inMedBin
#         meanAbsoDevi = cntSum*med + binByCnt*bin_width
        meanAbsoDevi = cntSum*med + binByCnt
        
        #debug
#         print len(beforeMed)
#         print 'MAD before median bin', beforeMed[0]*med + beforeMed[1]*bin_width
        
        sample_cnt = sorted_hist[med_bin_idx][1]
        sample_val= sorted_hist[ med_bin_idx][0] 
        sample_interval= bin_width*1.0/(sorted_hist[med_bin_idx][1]+1.0)
        
        for i in range(0, sample_cnt):
            sample_val= sample_val + sample_interval
            meanAbsoDevi= meanAbsoDevi + abs( sample_val-med)
        
        return (med,meanAbsoDevi)

#need further check
def split_onOneFeature_hist(values_hist,sumY_node,count_node,hist_node,meanAbsDevi_node):
    
    value_cnt= len(values_hist)
    
    leftY_sum=0
    left_count=0
    
    rightY_sum= sumY_node
    right_count= count_node
    
    bestSplitMetric=  meanAbsDevi_node 
    
    #this is a bit-record variable
    leftSplit_valueSet =0 
    leftSplit_count=0
    
    leftSplit_hist= {}
    rightSplit_hist={}
    
    bestLeft=(-1,-1)
    bestRight=(-1,-1)
    
    #sort feature values according to median 
    sorted_value_map=[]
    
    for i in values_hist.keys():
        
        currentVal_count= values_hist[i]['count_inFeatureValue']
        currentVal_hist= values_hist[i]['hist_inFeatureValue']
        currentVal_median = hist_meanAbsoluteDevi(currentVal_hist,currentVal_count)
        sorted_value_map.append( (currentVal_median[0],i) )
        
    sorted_value_map.sort()
    len_values= len(sorted_value_map)
    
    #debug
#     tmpcnt=0
#     for i in values_hist.keys():
#         tmpcnt= tmpcnt+ values_hist[i]['count_inFeatureValue']
#     if tmpcnt != count_node:
#         print '++++ problem feature count different from node count',count_node

    #debug
#     if len_values ==1:
#         print 'one value in a feature !!'


    
    node_hist_backup= copy.deepcopy(hist_node)
    
    # ini 
    current_feature_value= sorted_value_map[0][1]
    currentVal_count = values_hist[current_feature_value]['count_inFeatureValue']
    currentVal_hist=  values_hist[current_feature_value]['hist_inFeatureValue'] 
        
    leftSplit_hist=hist_merge( leftSplit_hist, currentVal_hist)
    rightSplit_hist= hist_minus( node_hist_backup, leftSplit_hist)
        
    leftSplit_count = leftSplit_count+currentVal_count
    left = hist_meanAbsoluteDevi( leftSplit_hist, leftSplit_count)
    right= hist_meanAbsoluteDevi( rightSplit_hist, (count_node - leftSplit_count))
    
    bestSplitMetric=1.0*leftSplit_count/count_node*left[1] + \
                               1.0*(count_node - leftSplit_count)/count_node*right[1]
             
    bestLeft=left
    bestRight=right        
    leftSplit_valueSet=leftSplit_valueSet+ (1<<current_feature_value)
        
    
    for k in range(1,len_values-1):
        
        current_feature_value= sorted_value_map[k][1]
        currentVal_count = values_hist[current_feature_value]['count_inFeatureValue']
        currentVal_hist=  values_hist[current_feature_value]['hist_inFeatureValue'] 
        
        leftSplit_hist=hist_merge( leftSplit_hist, currentVal_hist)
        rightSplit_hist= hist_minus( node_hist_backup, leftSplit_hist)
        
        leftSplit_count = leftSplit_count+currentVal_count
        left = hist_meanAbsoluteDevi( leftSplit_hist, leftSplit_count)
        right= hist_meanAbsoluteDevi( rightSplit_hist, (count_node - leftSplit_count))
        
        
        #debug
#         tmpcnt=0
#         for j in leftSplit_hist.keys():
#             tmpcnt = tmpcnt + leftSplit_hist[j]
#         if tmpcnt != leftSplit_count:
#             print '!!!! problem in left split', tmpcnt,leftSplit_count
            
#         tmpcnt=0
#         for j in rightSplit_hist.keys():
#             tmpcnt = tmpcnt + rightSplit_hist[j]
#         if tmpcnt != (count_node - leftSplit_count):
#             print '!!!! problem in right split',tmpcnt,(count_node - leftSplit_count) 
        
        leftMedian= left[0]
        leftMetric= left[1]
        rightMedian= right[0]
        rightMetric= right[1]
        
        #debug
#         if leftMedian<0 or rightMedian<0 or leftMetric <=0 or rightMetric <=0:
#             print '$$$$ problem in MAD calculation',leftMedian,rightMedian,leftMetric,rightMetric
        
        
        current_splitMetric=1.0*leftSplit_count/count_node*leftMetric + \
                               1.0*(count_node - leftSplit_count)/count_node*rightMetric
        
        if current_splitMetric < bestSplitMetric:
            
            bestSplitMetric=current_splitMetric  
            bestLeft=left
            bestRight=right
            #bit variable to record which value is chosen for the left child
            leftSplit_valueSet=leftSplit_valueSet+ (1<<current_feature_value)
    
    return (leftSplit_valueSet, [bestLeft[0], bestRight[0], bestSplitMetric  ] )


def find_bestSplit_hist(local_aggre_nodes, current_NumNodes, nodes_tree, \
                        nodes_tree_test):
    
    nodes_layer=[]
    current_idx_nodes =0
    
    #debug
    print 'debug:',len(local_aggre_nodes),current_NumNodes
    
    if local_aggre_nodes[ current_idx_nodes ][0] == -1:
        current_idx_nodes=1
    
    #debug 
    tmpnode_cnt=[]
    presplit=[] 
    
    for i in range(0, current_NumNodes):
        
       
        
        if local_aggre_nodes[ current_idx_nodes ][0] != i:
            nodes_layer.append( (-1,-1) )
            nodes_tree.append( (-1,-1) )
            continue
#       tmp_splitValue_oneFeature=[]
        
        tmp_sumY_node = local_aggre_nodes[i][1]['sumY'] 
        tmp_count_node = local_aggre_nodes[i][1]['count'] 
        tmp_node_hist=local_aggre_nodes[i][1]['hist'] 
    
        #debug
        tmpnode_cnt.append(tmp_count_node)
               
        best_splitMetric_feature= hist_meanAbsoluteDevi( tmp_node_hist, tmp_count_node)
        best_splitMetric_sofar= best_splitMetric_feature[1]
        preSplitMetric_node= best_splitMetric_feature[1]
        best_split=(-1,[-1,-1,-1])
        
        #debug
        presplit.append( best_splitMetric_feature[1]/100000 )
           
        best_split_feature=-1
        best_split_featureValueSet=-1
        
        
        #debug    
        flag=1
        for j in range(0, numFeatures):
            statisticToValues = local_aggre_nodes[i][1][j] 
            value_cnt= len(statisticToValues)
            if value_cnt >1:
                flag =0
                
        if flag ==1 :
            print 'debug: only one value for each feature in this node !!'
        
        
        
        for j in range(0,numFeatures):
                
            statisticToValues = local_aggre_nodes[i][1][j] 
            
#             (leftSplit_valueSet, [bestLeft[0], bestRight[0], bestSplitMetric  ] )
            split=split_onOneFeature_hist( statisticToValues, tmp_sumY_node,\
                                     tmp_count_node, tmp_node_hist, preSplitMetric_node)
    
            if split[1][2] < best_splitMetric_sofar:
                best_split=split
                best_splitMetric_sofar=split[1][2]
                best_split_feature = j
                best_split_featureValueSet= split[0]      
        
        # for one-layer point assignment
        nodes_layer.append( (best_split_feature,  best_split_featureValueSet  )  )
        
        # for whole-tree point assignment
        nodes_tree.append(  (best_split_feature,  best_split_featureValueSet  )      )
        
        # tree for predicting
        nodes_tree_test.append(best_split[1][0])
        nodes_tree_test.append(best_split[1][1])
        
        current_idx_nodes= current_idx_nodes+1
    
    #debug
#     print tmpnode_cnt
#     print presplit
    
    return nodes_layer
        
```

```python
# training and testing process

# dta_train = dta.sample(False, .7, 12345)
# dta_test = dta.sample(False, .3, 43243)
# dta_test = dta_test.filter(lambda line: line[0] <1000 )


# with outliers
tmpdta = dta.takeSample(False, 100000, 12243)
dta_train = sc.parallelize(tmpdta).cache().sample(False, .7, 12345)
dta_test = sc.parallelize(tmpdta).cache().sample(False, .3, 43243)

# without outliers
# tmpdta = dta.filter(lambda line: line[0]<1000).takeSample(False, 100000, 12243)
# dta_train = sc.parallelize(tmpdta).cache().sample(False, .7, 12345)
# dta_test =  sc.parallelize(tmpdta).cache().sample(False, .3, 43243)


maxdepth=10
numFeatures=3
bin_width=5

def search_nodeToData(features, tree):
    nodeNum=len(tree)   
    if nodeNum == 0:
        return 0;
    
    current_nodeIdx=0
    while current_nodeIdx< nodeNum:
        split_feature= tree[current_nodeIdx][0]
        split_valueSet= tree[current_nodeIdx][1]
        
        if split_feature==-1:
            return -1
        if (( 1<<features[split_feature]) &  split_valueSet) != 0 :
            current_nodeIdx=current_nodeIdx*2+1
        else:
            current_nodeIdx=current_nodeIdx*2+2        
    return current_nodeIdx - nodeNum

def tree_test_mapFunc(line):
    tmpnode = search_nodeToData(line[1:numFeatures+1], nodeSplits_tree)
    return (leaf_nodes[ tmpnode ]  - line[0])*(leaf_nodes[ tmpnode ]  - line[0])
    
def tree_test( testData_rdd ):
    err_rdd=testData_rdd.map( lambda line: tree_test_mapFunc(line))
    
    #debug
#     print err_rdd.take(20)
    
    err_sum = err_rdd.reduce( lambda a,b: a+b)
    return err_sum / testData_rdd.count()


#debug
def debug_tree_test_mapFunc(line):
    tmpnode = search_nodeToData(line[1:numFeatures+1], nodeSplits_tree)
    return (leaf_nodes[ tmpnode ], line[0])
    
def debug_tree_test( testData_rdd ):
    err_rdd=testData_rdd.map( lambda line: debug_tree_test_mapFunc(line))
#     print err_rdd.take(20)


#ini parameter on the cluster
currentNode_split_fromMaster = sc.broadcast([])
test_err = []
train_err = []
run_time=[]

for i in range(1,maxdepth):
    nodeSplits_tree=[]
    node_tree_test=[]
    currentNode_split_fromMaster = sc.broadcast([])
    
    start = time.time()
    
    for cur_depth in range(0,i):
        
        current_NumNodes= int( math.pow(2,  cur_depth)) 
        
        statis_partitions = dta_train.mapPartitions( partition_combiner_hist )
        aggre_nodes =statis_partitions.reduceByKey(lambda statis_partition_1, statis_partition_2: 
                              merge_parttion_combiner_hist(statis_partition_1,statis_partition_2 ))
        local_aggre_nodes= aggre_nodes.collect()
        cluster_end=time.time()
    
        nodeSplits_layer=find_bestSplit_hist(sorted(local_aggre_nodes,key= lambda val:val[0])\
                                    ,current_NumNodes,nodeSplits_tree,node_tree_test)
        #for whole tree
        currentNode_split_fromMaster = sc.broadcast(nodeSplits_tree)
        
    end = time.time()
    elapsed = end - start

    leaf_nodes = node_tree_test[ len(node_tree_test)-  (int)( math.pow(2,i)): len(node_tree_test)  ]
    
    tmp_test_err= tree_test( dta_test )
    tmp_train_err= tree_test( dta_train )
    
    #debug
    debug_tree_test( dta_test )
    
    test_err.append( tmp_test_err)
    train_err.append( tmp_train_err )
    run_time.append( elapsed )
    print "error at tree height", i,":",  tmp_test_err, tmp_train_err
    
```

```python
# training error inspection 
def error_partition_combiner_hist(list_dvAndfeatures):
    # each feature-value has a triple [cnt, mean, variance]
    
    nodes_dict={}
    tmpcnt=0
    
    for dvAndfeatures in list_dvAndfeatures:
        
        Y= dvAndfeatures[0]
        binId=math.floor(1.0* Y/bin_width )
        
        node= search_nodeToData(dvAndfeatures[1:numFeatures+1], \
                                currentNode_split_fromMaster.value)
        
        if node in nodes_dict.keys():
            
            # new added: sum of y in a node
            nodes_dict[node]['sumY'] = nodes_dict[node]['sumY']+Y 
            nodes_dict[node]['count'] = nodes_dict[node]['count']+1 
            nodes_dict[node]['sumYsquare'] = nodes_dict[node]['sumYsquare']+ Y*Y

            
            if binId in nodes_dict[node]['hist_node'].keys():
                nodes_dict[node]['hist_node'][binId]= nodes_dict[node]['hist_node'][binId]+1
            else:
                nodes_dict[node]['hist_node'].update( {binId:1}  )
            
        else:
            nodes_dict.update( {node: {}})  
            
            # new added: sum of y in a node
            nodes_dict[node].update( { 'sumY': Y} )
            nodes_dict[node].update( { 'count': 1} )
            nodes_dict[node].update( {'sumYsquare': Y*Y} )
            
            nodes_dict[node].update( {'hist_node':{} }  )
            nodes_dict[node]['hist_node'].update( {binId:1}  )
                
    return zip(nodes_dict.keys(), nodes_dict.values())


def error_merge_parttion_combiner_hist(nodeToFeatureToValue_1,nodeToFeatureToValue_2 ):
    
    # new added: sum of Y in a node
    nodeToFeatureToValue_1['sumY']= nodeToFeatureToValue_1['sumY']+ \
    nodeToFeatureToValue_2['sumY']
    
    nodeToFeatureToValue_1['count']= nodeToFeatureToValue_1['count']+\
    nodeToFeatureToValue_2['count']
    
    nodeToFeatureToValue_1['sumYsquare']= nodeToFeatureToValue_1['sumYsquare']+ \
    nodeToFeatureToValue_2['sumYsquare']
    
    
     for i in nodeToFeatureToValue_2['hist'].keys():
        if i in nodeToFeatureToValue_1['hist'].keys():
            nodeToFeatureToValue_1['hist'][i]= nodeToFeatureToValue_2['hist'][i]+\
            nodeToFeatureToValue_1['hist'][i]
        else:
            nodeToFeatureToValue_1['hist'].update( {i: nodeToFeatureToValue_2['hist'][i]})
    
    
    
 
    return  nodeToFeatureToValue_1
```
